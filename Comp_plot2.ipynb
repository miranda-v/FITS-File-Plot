```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hduN = fits.open('/users/mirandaviederman/Downloads/FITS Files/N_data/apStar-r8-2M16462176-0207594.fits')
hduN.info()
```

```python
hduS = fits.open('/users/mirandaviederman/Downloads/FITS Files/S_data/apStar-t9-2M16462176-0207594.fits')
hduS.info()
```

   ## Comparative Plot

#### North

```python
fluxesN = np.atleast_2d(hduN[1].data)
NexpN, NpixN = fluxesN.shape
wavelengthsN = 10. ** (hduN[0].header["CRVAL1"] + hduN[0].header["CDELT1"] * np.arange(NpixN))
```

#### South

```python
fluxesS = np.atleast_2d(hduS[1].data)
NexpS, NpixS = fluxesS.shape
wavelengthsS = 10. ** (hduS[0].header['CRVAL1'] + hduS[0].header['CDELT1'] * np.arange(NpixS))
```

### Plotting

```python
for exp in range(NexpN): 
    plt.step(wavelengthsN, 30*exp+fluxesN[exp], color="b", linewidth=0.3, alpha=.5)

for exp in range(NexpS): 
    plt.step(wavelengthsS, 30*exp+fluxesS[exp]+200, color="r", linewidth=0.3, alpha=.5)
    
plt.title('Comparative Plot of Northern and Southern Stellar Spectra Data')
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Ã…]')
plt.xlabel('Wavelength in Ã…ngstroms');

#plt.ylim(0,900);
#plt.xlim(15500,15700);
#plt.xlim(15700,15900);
#plt.xlim(16500,16750);
```
