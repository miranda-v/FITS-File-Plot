```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hduN = fits.open('/users/mirandaviederman/Downloads/FITS Files/N_data/apStar-r8-2M16471235-0153543.fits')
hduN[0].header
```

   ## Comparative Plot

### So far only Northern data is accessible

```python
fluxesN = np.atleast_2d(hduN[1].data)
NexpN, NpixN = fluxesN.shape
wavelengthsN = 10. ** (hduN[0].header["CRVAL1"] + hduN[0].header["CDELT1"] * np.arange(NpixN))
plt.clf()
for exp in range(NexpN):
    plt.step(wavelengthsN, 50*exp+fluxesN[exp], label = 'Exposure number: '+str(exp+1), linewidth=.2, alpha=.7)
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Å]')
plt.xlabel("Wavelength in Ångstroms")
plt.title('Comparative Plot of Northern and Southern Stellar Spectra Data')
plt.legend(loc = 'best', ncol=2, fontsize=8);
#plt.xlim(15500,15600);
```
