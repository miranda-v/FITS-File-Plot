```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hduN = fits.open('/users/mirandaviederman/Downloads/FITS Files/N_data/apStar-r8-2M16471235-0153543.fits')
hduN.info()
```

```python
hduS = fits.open('/users/mirandaviederman/Downloads/FITS Files/S_data/apStar-t9-2M16471235-0153543.fits')
hduS.info()
```

   ## Comparative Plot

#### North -- APO

```python
fluxesN = np.atleast_2d(hduN[1].data)
NexpN, NpixN = fluxesN.shape
wavelengthsN = 10. ** (hduN[0].header["CRVAL1"] + hduN[0].header["CDELT1"] * np.arange(NpixN))
```

#### South -- LCO

```python
fluxesS = np.atleast_2d(hduS[1].data)
NexpS, NpixS = fluxesS.shape
wavelengthsS = 10. ** (hduS[0].header["CRVAL1"] + hduS[0].header["CDELT1"] * np.arange(NpixS))
```

### Plotting

```python
for exp in range(NexpN):
    if exp==0:
        plt.step(wavelengthsN, 50*exp+fluxesN[exp], color='r', linewidth=.2, alpha=.5, label='Apache Point Observatory')
    else:
        plt.step(wavelengthsN, 50*exp+fluxesN[exp], color='r', linewidth=.2, alpha=.5, label='_nolegend_')
    
for exp in range(NexpS):
    if exp==0:
        plt.step(wavelengthsS, 50*exp+fluxesS[exp], color='b', linewidth=.2, alpha=.5, label='Las Campanas Observatory')
    else:
        plt.step(wavelengthsS, 50*exp+fluxesS[exp], color='b', linewidth=.2, alpha=.5, label='_nolegend_')
    
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Ã…]')
plt.xlabel("Wavelength in Ã…ngstroms")
plt.title('Comparative Plot of Northern and Southern Stellar Spectra Data')
plt.legend(loc = 'best', ncol=2, fontsize=8);
#plt.xlim(15500,15600);
```
