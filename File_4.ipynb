```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hdu4 = fits.open('/users/mirandaviederman/Downloads/FITS Files/apStar-r8-2M05580827+1345138.fits')
hdu4.info()
```

## Plot

```python
fluxes = np.atleast_2d(hdu4[1].data)
Nexp, Npix = fluxes.shape
wavelengths = 10. ** (hdu4[0].header["CRVAL1"] + hdu4[0].header["CDELT1"] * np.arange(Npix))
plt.clf()
for exp in range(Nexp):
    plt.step(wavelengths, 100*exp+fluxes[exp], label = 'Exposure number: '+str(exp+1), linewidth = .5, alpha = .7)
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Ã…]')
plt.xlabel("Wavelength in Ã…ngstroms")
plt.legend(loc='best', ncol=2, fontsize=8)
plt.title('APOGEE ID 2M05580827+1345138');
#plt.xlim(15500, 15700);
```
