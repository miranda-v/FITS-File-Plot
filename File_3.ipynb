```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hdu3 = fits.open('/users/mirandaviederman/Downloads/FITS Files/apStar-r8-2M08463782+5729103.fits')
hdu3.info()
```

## Plot

```python
fluxes = np.atleast_2d(hdu3[1].data)
Nexp, Npix = fluxes.shape
wavelengths = 10. ** (hdu3[0].header["CRVAL1"] + hdu3[0].header["CDELT1"] * np.arange(Npix))
plt.clf()
for exp in range(Nexp):
    plt.step(wavelengths, 150*exp+fluxes[exp], label = 'Exposure number: '+str(exp+1), linewidth = .3, alpha=0.65)
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Ã…]')
plt.xlabel("Wavelength in Ã…ngstroms")
plt.title('APOGEE ID 2M08463782+5729103')
plt.legend();
#plt.xlim(16250, 16450);
```
