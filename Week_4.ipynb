```python
hdu1 = fits.open('/users/mirandaviederman/Downloads/FITS Files/apStar-r8-2M13384267+2723430.fits')
hdu1.info()
```

## Plots

```python
fluxes = np.atleast_2d(hdu1[1].data)
Nexp, Npix = fluxes.shape
wavelengths = 10. ** (hdu1[0].header["CRVAL1"] + hdu1[0].header["CDELT1"] * np.arange(Npix))
plt.clf()
for exp in range(Nexp):
    plt.step(wavelengths, fluxes[exp], label = 'APOGEE ID 2M13384267+2723430.fits')
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/]')
plt.xlabel("Wavelength in ngstroms")
plt.legend();
```

### 20-ngstrom portion of plot

```python
hdu1 = fits.open('/users/mirandaviederman/Downloads/FITS Files/apStar-r8-2M13072813+1653414.fits')
fluxes = np.atleast_2d(hdu1[1].data)
Nexp, Npix = fluxes.shape
wavelengths = 10. ** (hdu1[0].header["CRVAL1"] + hdu1[0].header["CDELT1"] * np.arange(Npix))
plt.clf()
for exp in range(Nexp):
    plt.step(wavelengths, fluxes[exp] + 100 * exp, alpha=0.25)
plt.ylabel("flux")
plt.xlabel("wavelength")
plt.xlim(15200, 15220)
```
