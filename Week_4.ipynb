```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
import astropy
from astropy.io import fits
```

```python
hdu1 = fits.open('/users/mirandaviederman/Downloads/FITS Files/apStar-r8-2M13384267+2723430.fits')
hdu1.info()
```

## Plot

```python
fluxes = np.atleast_2d(hdu1[1].data)
Nexp, Npix = fluxes.shape
wavelengths = 10. ** (hdu1[0].header["CRVAL1"] + hdu1[0].header["CDELT1"] * np.arange(Npix))
plt.clf()
for exp in range(Nexp):
    plt.step(wavelengths, fluxes[exp], label = 'APOGEE ID 2M13384267+2723430', linewidth = .5)
plt.ylabel(r'Flux [$10^{-17}$ erg/cm$^2$/s/Ã…]')
plt.xlabel("Wavelength in Ã…ngstroms")
plt.legend();
#plt.xlim(16000, 16200);
```
